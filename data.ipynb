```python
#Method to retrieve list of songs from a playlist on Spotify

def get_songList(playlist):
    # get information about playlist
    result = sp.playlist_items(playlist)
    songs = result['items']
    print(songs)

    #access code for all songs
    while result['next']:
        result = sp.next(result)
        songs.extend(result['items'])

    # retrieve number of indices in list
    elements = len(songs)

    track_uris = []

    #create list of all song uris in playlist
    for x in range(elements):
        song = songs[x]
        temp = song['track']['uri'].split(":")
        track_uris.append(temp[2])
    return track_uris
```

```python
# base URL of all Spotify API endpoints
BASE_URL = 'https://api.spotify.com/v1/'
import pandas as pd
import numpy as np
import requests
import re
import time
```

```python
#create initial data frame from song list with Spotify song metrics
def get_spotFrame(songList):

    # create blank dictionary to store audio features
    feature_dict = {}

    # convert track_uri column to an iterable list
    #track_uris = ["6m9qPYXmhge2QhBLfFKnVF"]

    # loop through track URIs and pull audio features using the API,
    # store all these in a dictionary
    count = 0
    for t_uri in songList:
        
        time.sleep(1)
        feature_dict[t_uri] = {'popularity': 0,
                               'name': 0,
                               'danceability': 0,
                               'energy': 0,
                               'speechiness': 0,
                               'instrumentalness': 0,
                               'tempo': 0}

        r = requests.get(BASE_URL + 'tracks/' + t_uri, headers=headers)
        r = r.json()
        feature_dict[t_uri]['popularity'] = r['popularity']
        name = r['name']
        if "-" in r['name']:
            dex = r['name'].index("-")
            name = r['name'][0:dex]
        if "(" in r['name']:
            dex = r['name'].index("(")
            name = r['name'][0:dex]
        feature_dict[t_uri]['name'] = name

        s = requests.get(BASE_URL + 'audio-features/' + t_uri, headers=headers)
        s = s.json()
        feature_dict[t_uri]['danceability'] = s['danceability']
        feature_dict[t_uri]['energy'] = s['energy']
        feature_dict[t_uri]['speechiness'] = s['speechiness']
        feature_dict[t_uri]['instrumentalness'] = s['instrumentalness']
        feature_dict[t_uri]['tempo'] = s['tempo']
        count += 1
    #print(feature_dict)
    # convert dictionary into dataframe with track_uri as the first column
    df_features = pd.DataFrame.from_dict(feature_dict, orient='index')
    df_features.insert(0, 'track_uri', df_features.index)
    df_features.reset_index(inplace=True, drop=True)

    return df_features
```
